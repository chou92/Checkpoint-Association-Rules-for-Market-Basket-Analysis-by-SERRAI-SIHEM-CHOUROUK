```python
import pandas as pd
import numpy as np
from pandas import DataFrame,Series
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
Market_file = "Market_Basket_OptimisationL"
Market_data = pd.read_csv("Market_Basket_Optimisation.csv", sep=',')
Market_data
```

```python
Market_data.shape
```

```python
transactions=[]
for i in range(0,7500):
    transactions.append([str(Market_data.values[i,j]) for j in range(0,20)])
```

```python
transactions[:2]
```

## Training the apriori model

```python
pip install apyori
```

```python
from apyori import apriori
```

```python
from apyori import apriori
rules=apriori(transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(rules)
results
```

## Visualization of Results

```python
lhs,rhs,support,confidence,lift=[],[],[],[],[]
for result in results:
    lhs.append(tuple(result[2][0][0])[0])
    rhs.append(tuple(result[2][0][1])[0])
    support.append(result[1])
    confidence.append(result[2][0][2])
    lift.append(result[2][0][3])
```

```python
columns=list(zip(lhs,rhs,support,confidence,lift))
```

```python
result_df=pd.DataFrame(columns)
result_df.columns=['Item1','Item2','Support','Confidence','Lift']
result_df
```

```python
result_df=result_df.sort_values('Lift',ascending=False)
result_df
```

```python
print("There are {} Relation derived.".format(len(result_df)))
```
